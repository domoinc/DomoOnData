```python
import numpy as np
import pandas as pd
import domojupyter as domo
import csv
import requests
import re
import time
import json
import xml.etree.ElementTree as ET


report='.CPI.TOT+FOOD+TOT_FOODENRG.AGRWTH.M'
dataset_identifier='DP_LIVE'
agency_name='OECD'
#report='.CPIFORECAST.TOT.AGRWTH.Q'
#report='AUS+AUT+BEL+CAN+CHL+COL+CRI+CZE+DNK+EST+FIN+FRA+DEU+GRC+HUN+ISL+IRL+ISR+ITA+JPN+KOR+LVA+LTU+LUX+MEX+NLD+NZL+NOR+POL+PRT+SVK+SVN+ESP+SWE+CHE+TUR+GBR+USA+EA19+EU28+EU27_2020+G7M+G20+OECDE+OECD+WLD+NMEC+AFG+ALB+DZA+ASM+AND+AGO+AIA+ATA+ATG+ARG+ARM+ABW+AZE+BHS+BHR+BGD+BRB+BLR+BLZ+BEN+BMU+BTN+BOL+BIH+BES+BWA+BRA+VGB+BRN+BGR+BFA+BDI+KHM+CMR+CPV+CYM+CAF+TCD+CHN+CXR+COM+COG+COD+COK+CIV+HRV+CUB+CYP+CUW+DJI+DMA+DOM+ECU+EGY+SLV+GNQ+ERI+ETH+FRO+FJI+GAB+GMB+GEO+GHA+GTM+GIN+GNB+GUY+HTI+HND+IND+IDN+IRN+IRQ+JAM+JEY+JOR+KAZ+KEN+KIR+KWT+KGZ+LAO+LBN+LSO+LBR+LBY+LIE+MAF+MDG+MWI+MYS+MDV+MLI+MLT+MHL+MTQ+MRT+MUS+MYT+FSM+MDA+MCO+MNG+MNE+MSR+MAR+MOZ+MMR+NAM+NRU+ANT+NCL+NIC+NER+MKD+NGA+NIU+NFK+MNP+OMN+PAK+PLW+PAN+PNG+PRY+PER+PHL+PRI+QAT+ROU+RUS+RWA+SHN+BLM+KNA+LCA+SXM+SPM+VCT+WSM+SMR+STP+SAU+SEN+SRB+SYC+SLE+SGP+SLB+SOM+ZAF+LKA+SDN+SUR+SSD+SWZ+SYR+TJK+TZA+THA+TLS+TGO+TKL+TON+TTO+TUN+TKM+TCA+TUV+UGA+UKR+ARE+URY+UZB+VUT+VEN+VNM+VIR+WLF+YEM+ZMB+ZWE+GRPS+LAC+MENA+ASEAN+BRIICS+OECDAM+OECDAO+EECCA+CPT.ENVRESPROD+CO2+CO2_PBPROD+CO2_PBEMCAP+CO2_PBEM00+CO2_PBEM+CO2_DBPROD+CO2_DBPROD_NNDI+CO2_DBEMCAP+CO2_DBEM00+CO2_DBEM+CO2_INTPROD+CO2_AIRTRACAP+CO2_AIRTRAGDP+NRG+NRG_PROD+NRG_INT+NRG_I00+NRGS+RE_TPES+RE_NRG+NRGC_AGR+NRGC_SER+NRGC_IND+NRGC_TRA+NRGC_OTH+RE_TPES_EBIOM+MAT+DMC_PROD+DMC_BIO+DMC_MIN+DMC_MET+MWAS_INT+MWAS_INC+MWAS_RECO+MWAS_LANDF+NBAL_HA+PBAL_HA+EAMFP+EAMFP_EAMFPG+EAMFP_NKG+EAMFP_APAG+NATBASE+WATER+WATER_STPC+WATER_FWCAP+WATER_FW_TR+WATER_FW_TIRR+SW_PERMWAT+SW_SEASWAT+SW_PERMTONOT+SW_PERMTOSEAS+SW_NOTTOPERM+SW_SEASTOPERM+LAND+LC_NAT+LC_BAR+LC_CRO+LC_URB+LC_WAT+LCC_NATLOSS92+LCC_NATLOSS04+LCC_NATGAIN92+LCC_NATGAIN04+LCC_NATTOCROP+LCC_NATTOARTIF+LCC_CROPTOARTIF+BA_BUILTA+BA_BUILTA_CAP+BA_NEWBUILTA90+BA_NEWBUILTA00+IRRIG_AGLAND+FOR+FOR_STOCK+FOR_INTUSE+FOR_LTMGMT+FOR_NAT+FOR_IF_EXTENT+FOR_IF_PCLOSS2000+WLIFE+WLIFE_MA+WLIFE_BI+WLIFE_PL+BIRD_FARM+BIRD_FOR+PEST_AGLAND+ATMOSPHERE+TEMPCHANGE5180+ENVQOL+RENV+PM_PWM+PM_SPEX10+PM_SPEX35+PM_MOR+PM_SC+O3_MOR+O3_SC+PB_MOR+PB_SC+RN_MOR+RN_SC+ASEW+DRIN_SPOP+SANI_SPOP+ASEW_POP+ASEW_PWT+ASEW_SWT+ASEW_TWT+ECOPOL+TECHPAT_PAT+GPAT_DE_AT+GPAT_DE_RTA+GPAT_DE_AI+GPAT_DE_CAP+TECHPAT_RD+ENVRD_GBAORD+ENVRD_GDP+RERD_ERD+ERD_GDP+FFRD_ERD+EGS+EPEA_GDP+IFF_ODA+ODA_ENV+ODA_BIO+ODA_CCMIT+ODA_CCADP+ODA_DES+ODA_RE+ODA_WATER+ODA_ENVSEC+ODA_GNI+ENVTAXTRANS+ENVTAX_GDP+ENVTAX_TR+ENVTAX_NRG+ENVTAX_VEH+ECR_PC30UP+ECR_PC60UP+ECR_PC120UP+FTAX_DIE+FPRICE_DIE+FTAX_PET+FPRICE_PET+EPRICE_RES+EPRICE_IND+FIT_SOLAR+FIT_WIND+CSE_TOTT+CSE_ENET+CSE_FFS+PSE_FFS+GSSE_FFS+PET_FFS+COAL_FFS+NATG_FFS+ELEC_FFS+FFS_TTAX+REGMNG+PA_TERRESTRIAL+PA_MARINE+SOCECO+ECO+GDP_R00+GDP_R+AGRGDP_PC+INDGDP_PC+SRVGDP_PC+LTAX_GDP+LTAX_TTAX+GDP_RCAP+XR+PPP+DEF+SOC+POP+POP_A0014+POP_A1564+POP_A65UP+POP_SWOMEN+POP_FERTILITY+POP_LIFEEXP+POP_NETMIGR+POPDEN'
#dataset_identifier='GREEN_GROWTH'
#agency_name='all'
parameters='startPeriod=1900-01'
outputfilename='oecd'
outputdatasetname='OECD_CPI'

FinalHeader=[]
TEMP_Rows=[]

f=open(''+outputfilename+'.csv', 'w')
writer = csv.writer(f)

def get_xml():
	printheadercount=0
	URL = "https://stats.oecd.org/restsdmx/sdmx.ashx/GetData/"+dataset_identifier+"/"+report+"/"+agency_name+"?json-lang=en&dimensionAtObservation=allDimensions&"+parameters+"&format=compact_v2"
	print(URL)
	response = requests.get(URL)
	responsetxt=str(response.text)
	responsetxt=responsetxt.replace('ï»¿','')
	#print(responsetxt)
	root = ET.fromstring(responsetxt)
	for child in root[1]:
		header=[]
		row=[]
		if len(child.attrib)>0:
			for x in child.attrib:
				#print(x)
				header.append(x)
				#print(child.get(x))
				row.append(child.get(x))
			for grandchild in child:
				grandchildheader=[]
				grandchildrow=[]
				for y in grandchild.attrib:
					#print(y)
					grandchildheader.append(y)
					#print(grandchild.get(y))
					grandchildrow.append(grandchild.get(y))


				fullheader=[]
				fullrow=[]
				for h in header:
					if h not in FinalHeader:
						FinalHeader.append(h)
						#print(FinalHeader)
					fullheader.append(h)
				
				for gh in grandchildheader:
					#print(gh)
					if gh not in FinalHeader:
						FinalHeader.append(gh)
					fullheader.append(gh)
				#print(fullheader)
				for r in row:
					fullrow.append(r)
				for gr in grandchildrow:
					fullrow.append(gr)
				temprow={}
				for fr in range(len(fullrow)+1):
					temprow.update({fullheader[fr-1]:fullrow[fr-1]})
				#print(temprow)
				TEMP_Rows.append(temprow)
				#print(fullrow)


	if printheadercount==0:
		print(FinalHeader)
		writer.writerow(FinalHeader)
		printheadercount=1

	for x in range(len(TEMP_Rows)):
		tempfinalrowinfo=[]
		for y in FinalHeader:
			if y in TEMP_Rows[x]:
				tempfinalrowinfo.append(TEMP_Rows[x][y])
			else:
				tempfinalrowinfo.append('')
		writer.writerow(tempfinalrowinfo)


get_xml()

f.close()



df=pd.read_csv('oecd.csv', delimiter = ',')
print(df)

domo.write_dataframe(df,'OECD_CPI')
```
